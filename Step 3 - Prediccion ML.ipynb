```python
# 1. Asegurar tipos de datos para el cruce
df_sales['ITEM_CODE'] = df_sales['ITEM_CODE'].astype(str)
df_product['ITEM'] = df_product['ITEM'].astype(str)

# 2. Realizar los Merges seleccionando solo las columnas deseadas
# Primero unimos Ventas con los detalles de Producto
df_temp = pd.merge(
    df_sales, 
    df_product[['ITEM', 'BRAND', 'ITEM_DESCRIPTION']], 
    left_on='ITEM_CODE', 
    right_on='ITEM', 
    how='left'
)

# Luego unimos con Calendario para traer la DATE
df_final = pd.merge(
    df_temp, 
    df_calendar[['WEEK', 'DATE']], 
    on='WEEK', 
    how='left'
)

# 3. Selección y Orden de columnas (El "Ponchado" final)
columnas_deseadas = [
    'WEEK', 
    'DATE', 
    'ITEM_CODE', 
    'BRAND', 
    'ITEM_DESCRIPTION', 
    'TOTAL_UNIT_SALES', 
    'TOTAL_VALUE_SALES', 
    'TOTAL_UNIT_AVG_WEEKLY_SALES', 
    'REGION'
]

df_final = df_final[columnas_deseadas]

# 4. Convertir DATE a formato fecha y filtrar años 2022-2023
df_final['DATE'] = pd.to_datetime(df_final['DATE'])
df_final = df_final[df_final['DATE'].dt.year.isin([2022, 2023])]

print("✅ DataFrame Final consolidado con éxito.")
print(f"Dimensiones: {df_final.shape}")
df_final.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Filtrar solo Cloralex
df_cloralex = df_final[df_final['BRAND'] == 'CLORALEX'].copy()

# 2. Asegurar que DATE sea formato fecha
df_cloralex['DATE'] = pd.to_datetime(df_cloralex['DATE'])

# 3. Agrupar por fecha para tener la venta total semanal de la marca
df_ml = df_cloralex.groupby('DATE')['TOTAL_VALUE_SALES'].sum().reset_index()
df_ml = df_ml.sort_values('DATE')

# 4. Crear variables para la Regresión Múltiple (Tendencia y Mes)
df_ml['Trend'] = np.arange(len(df_ml))
df_ml['Month'] = df_ml['DATE'].dt.month

print(f"Datos listos: {len(df_ml)} semanas registradas.")
```

```python
df_cloralex
```

```python
# --- MODELO 1: Regresión Lineal Múltiple ---
X = df_ml[['Trend', 'Month']]
y = df_ml['TOTAL_VALUE_SALES']

model_lr = LinearRegression()
model_lr.fit(X, y)
df_ml['Pred_Linear'] = model_lr.predict(X)

# --- MODELO 2: Series de Tiempo ARIMA ---
# Usamos un orden (5,1,0) que suele ser robusto para datos semanales
model_arima = ARIMA(df_ml['TOTAL_VALUE_SALES'], order=(5,1,0))
model_arima_fit = model_arima.fit()
df_ml['Pred_ARIMA'] = model_arima_fit.fittedvalues
```

```python
# Calculamos métricas para ARIMA (que suele ser el más preciso en series de tiempo)
mae_arima = mean_absolute_error(df_ml['TOTAL_VALUE_SALES'], df_ml['Pred_ARIMA'])
mse_arima = mean_squared_error(df_ml['TOTAL_VALUE_SALES'], df_ml['Pred_ARIMA'])

print(f"--- VALIDACIÓN DEL MODELO ---")
print(f"Error Absoluto Medio (MAE): ${mae_arima:,.2f}")
print(f"Error Cuadrático Medio (MSE): {mse_arima:,.2f}")
```

```python
# Pronóstico con ARIMA para las próximas 12 semanas
forecast_steps = 12
forecast = model_arima_fit.forecast(steps=forecast_steps)

# Crear fechas futuras
last_date = df_ml['DATE'].max()
future_dates = pd.date_range(start=last_date + pd.Timedelta(weeks=1), periods=forecast_steps, freq='W')

# Gráfico Final
plt.figure(figsize=(12, 6))
plt.plot(df_ml['DATE'], df_ml['TOTAL_VALUE_SALES'], label='Ventas Reales', color='blue')
plt.plot(df_ml['DATE'], df_ml['Pred_ARIMA'], label='Ajuste del Modelo (Pasado)', color='green', linestyle='--')
plt.plot(future_dates, forecast, label='Predicción Futura (3 meses)', color='red', marker='o')
plt.title('Predicción de Ventas Cloralex: Real vs Modelo vs Futuro')
plt.xlabel('Fecha')
plt.ylabel('Ventas en Valor')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Para ARIMA, el summary se obtiene directamente del objeto fit
print(model_arima_fit.summary())
```

```python
# 1. Obtener la predicción y el intervalo de confianza para el futuro
forecast_object = model_arima_fit.get_forecast(steps=12)
forecast_mean = forecast_object.predicted_mean
forecast_ci = forecast_object.conf_int() # Aquí están los rangos (mínimo y máximo)

# 2. Graficar
plt.figure(figsize=(12, 6))

# Ventas Reales
plt.plot(df_ml['DATE'], df_ml['TOTAL_VALUE_SALES'], label='Ventas Reales', color='blue', linewidth=2)

# Predicción Futura
plt.plot(future_dates, forecast_mean, label='Predicción ARIMA', color='red', linestyle='--')

# Rango de Error (Sombra)
plt.fill_between(future_dates, 
                 forecast_ci.iloc[:, 0], 
                 forecast_ci.iloc[:, 1], 
                 color='pink', alpha=0.3, label='Intervalo de Confianza (95%)')

plt.title('Pronóstico de Ventas Cloralex con Rango de Incertidumbre')
plt.xlabel('Fecha')
plt.ylabel('Ventas en Valor')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
residuals = model_arima_fit.resid
plt.figure(figsize=(10, 4))
plt.plot(residuals, color='purple')
plt.title('Residuos del Modelo (Errores de Predicción)')
plt.axhline(0, color='black', linestyle='--')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Configurar la serie de tiempo (usando la venta total de Cloralex)
# Es importante definir el 'period' (52 semanas si es anual, o lo que dure tu ciclo)
analysis = df_ml.set_index('DATE')['TOTAL_VALUE_SALES']
decomposition = seasonal_decompose(analysis, model='additive', period=12) 

# 2. Graficar los 4 displays
plt.rcParams.update({'figure.figsize': (12, 10)})
decomposition.plot()
plt.suptitle('Descomposición Estacional de Ventas Cloralex', fontsize=16, y=1.02)
plt.show()
```

```python
# 1. Cortamos el pasado: Solo nos quedamos con datos de Abril 2023 en adelante
df_inercia = df_ml[df_ml['DATE'] >= '2023-04-01'].copy()

# 2. Re-entrenamos el modelo ARIMA solo con esta "nueva realidad"
# Nota: Al ser menos datos, ajustamos el orden a (2,1,0) para evitar sobreajuste
model_nueva_inercia = ARIMA(df_inercia['TOTAL_VALUE_SALES'], order=(2,1,0))
fit_inercia = model_nueva_inercia.fit()

# 3. Pronóstico para los próximos 3 meses (12 semanas)
forecast_steps = 12
forecast_mean = fit_inercia.get_forecast(steps=forecast_steps).predicted_mean
forecast_ci = fit_inercia.get_forecast(steps=forecast_steps).conf_int()

# 4. Crear las fechas futuras
last_date = df_inercia['DATE'].max()
future_dates = pd.date_range(start=last_date + pd.Timedelta(weeks=1), periods=forecast_steps, freq='W')

# 5. Graficar la "Nueva Realidad"
plt.figure(figsize=(12, 6))

# Graficamos el histórico corto (Abril - Julio)
plt.plot(df_inercia['DATE'], df_inercia['TOTAL_VALUE_SALES'], label='Nueva Inercia (Abril-Julio 2023)', color='#1f77b4', marker='o', linewidth=2)

# Graficamos la proyección
plt.plot(future_dates, forecast_mean, label='Proyección de Crecimiento', color='red', linestyle='--', linewidth=3)

# Sombra de incertidumbre
plt.fill_between(future_dates, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.1)

plt.title('Modelo de Predicción: Enfoque en Nueva Inercia de Mercado (Post-Abril 2023)', fontsize=14)
plt.xlabel('Fecha')
plt.ylabel('Ventas en Valor')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
